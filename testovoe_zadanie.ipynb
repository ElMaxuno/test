```python
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
import pandas as pd
```

####решим задачу при помощи симуляций в python, в течении 10000 раз подбрасывая монету N  раз и затем посмотрим в скольки из 10000 случаев 60% и более орлов.

```python
NUM_experiments =10000
results = []
for LEN_N in tqdm(range(1,150)):
  means = []
  for experiment in range(NUM_experiments):
    sum = []
    for i in range(LEN_N):
      if np.random.uniform()>.5:
        sum.append(0)
      else:
        sum.append(1)
    means.append(np.mean(sum))
  results.append(np.array(means))
```

```python
y = [np.sum(i>.6)/NUM_experiments for  i in results]
iter = np.arange(1,150)
```

####Зависимость искомой вероятности от N

```python
plt.figure(figsize=(10,8))
plt.plot(iter ,y)
plt.xlabel('N')
plt.ylabel('вероятность, получить  60% и более решек')
```

```python
data = pd.DataFrame()
```

```python
data['N'] = iter
data['proba'] = y
```

###будем считать монету нечестной, если вероятность выпадания такого случая меньше 0.05
